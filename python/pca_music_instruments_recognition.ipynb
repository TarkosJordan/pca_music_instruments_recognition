```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
project_dir = 'C:\\Users\\tarsi\\OneDrive\\Documentos\\Python\\pca_music_instruments_recognition'
audio_base_path = os.path.join(project_dir, 'base_songs')

print(audio_base_path)

for arquivo in os.listdir(audio_base_path):
    nome, extensao = os.path.splitext(arquivo)

    if extensao == '.mp3':
        # Carregando o arquivo de udio
        audio_path = os.path.join(audio_base_path, arquivo)

        y, sr = librosa.load(audio_path)

        # Aplicando o STFT
        D = librosa.stft(y, n_fft=4096, hop_length=512, win_length=2048, center=True)

        # Obtendo o espectrograma em dB
        D_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

        # Plotando o espectrograma
        plt.figure(figsize=(12, 6))
        librosa.display.specshow(D_db, sr=sr, x_axis='time', y_axis='log')
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'Espectrograma do udio: {nome}')
        plt.show()
```
